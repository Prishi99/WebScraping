```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.5",
    "Referer": "https://www.ambitionbox.com",
    "Connection": "keep-alive"
}
```

```python
import requests

session = requests.Session()
session.headers.update(headers)

url = 'https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=2'
webpage = session.get(url)

print(webpage.text)
```

```python
soup = BeautifulSoup(webpage.text, 'lxml')
```

```python
for i in soup.find_all('span', class_='companyCardWrapper__ActionTitle'):
  print(i.text.strip())
```

```python
company = soup.find_all('div', class_='companyCardWrapper')
```

```python
len(company)
```

```python
name=[]
rating=[]
review=[]
location=[]
for i in company:
  name.append(i.find('h2').text.strip())
  rating.append(i.find('div', class_='rating_star_container').text.strip())
  review.append(i.find('span', class_='companyCardWrapper__ActionCount').text.strip())
  location.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip())

```

```python
company_info = []

for i in company:
    label = i.find('span', class_ ="companyCardWrapper__ratingHeader--high")
    rating = i.find('span', class_='companyCardWrapper__ratingValues')

    if label and rating:
        combined = f"{label.text.strip()}: {rating.text.strip()}"
        company_info.append(combined)
    else:
        company_info.append("Not available")
```
